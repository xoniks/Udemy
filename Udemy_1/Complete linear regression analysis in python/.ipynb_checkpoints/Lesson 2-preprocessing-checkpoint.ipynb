Treating percentiles we have three ways:
    1. capping and flooring
        indentify outliers below P1 and bigger than P99
        Then multiply p99*3 and p1*0.3
    2.exponential smoothing
    3.sigma approach
 

```python
#to indentify percentiles we use numpy packages built-in functions percentiles
np.percentile(df.n_hot_rooms,[99])
```

```python
#to convert the result from percentile that is an array to number
np.percentile(df.n_hot_rooms,[99])[0]
```

```python
uv=np.percentile(df.n_hot_rooms,[99])[0]
```

```python
#to sort the percentiles values more than the 99th percentile
df[(df.n_hot_rooms>uv)]
```

```python
#for capping and flooring we need to classify the df.n_hot_rooms that are more than 3*uv and equalize with 3*uv
df.n_hot_rooms[(df.n_hot_rooms)>3*uv]=3*uv
```

```python
sns.jointplot(x='rainfall',y='price',data=df)
```

```python
np.percentile(df.rainfall,[1])[0]
```

```python
lv=np.percentile(df.rainfall,[1])[0]
```

```python
#to show all the values that are smaller than fisrt percentile
df[(df.rainfall<lv)]
```

```python
#after we find the outliers for smaller values than first percentile we eaqulize it with 0.3 times first percentiles values because of flooring and capping method
df[df.rainfall<0.3*lv]=0.3*lv
```

```python
#the relation between crime rate and price is exponential not linear, so to treat this we need to use different method
#in this cases we need to transform first then check again if the outliers are present or not
sns.jointplot(x='crime_rate',y='price',data=df)
```
